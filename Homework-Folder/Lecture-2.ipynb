```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset from Regensburg Pediatric Appendicitis
pediatric = fetch_ucirepo(id=938) 
  
# data (as pandas dataframes) 
X = pediatric.data.features 
y = pediatric.data.targets  
  
# variable information 
print(pediatric.variables) 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Plot the distribution of weight
plt.figure(figsize=(10, 6))
sns.histplot(X['Weight'], bins=20, kde=True, color='orange')
plt.title('Weight Distribution')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of lymph nodes
plt.figure(figsize=(10, 6))
sns.countplot(x='Lymph_Nodes_Location', data=X, palette='viridis')
plt.title('Distribution of Lymph_Nodes')
plt.xlabel('Lymph_Nodes Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Plot the relationship between Sex and Length_of_Stay
plt.figure(figsize=(12, 8))
sns.countplot(x='Sex', hue='Length_of_Stay', data=X, palette='Set2')
plt.title('Sex vs Length_of_Stay')
plt.xlabel('Sex Classification')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Length_of_Stay', labels=['M', 'F'])
plt.show()

# Plot the distribution of Stool
plt.figure(figsize=(12, 8))
occupation_counts = X['Stool'].value_counts()
plt.pie(occupation_counts, labels=occupation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Stool')
plt.show()


# Plot the distribution of Neutrophilia
plt.figure(figsize=(10, 6))
sns.countplot(x='Neutrophilia', data=X, palette='muted')
plt.title('Distribution of Neutrophilia')
plt.xlabel('Neutrophilia Classification')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Plot the relationship between sex and weight
plt.figure(figsize=(12, 8))
sns.violinplot(x='Sex', y='Weight', data=X, palette='pastel')
plt.title('Sex Vs Weight')
plt.xlabel('Sex')
plt.ylabel('Weight')
plt.xticks(rotation=45)
plt.show()
```
